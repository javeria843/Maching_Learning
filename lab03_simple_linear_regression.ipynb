```python
# Task 1: Simple Linear Regression on Diabetes Dataset (BMI vs Disease Progression)

import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# --------------------------
# Load Dataset
# --------------------------
diabetes = load_diabetes()
X = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
y = diabetes.target

# Select only BMI feature
X_bmi = X[["bmi"]]

st.title("📊 Task 1: Simple Linear Regression (BMI vs Disease Progression)")

st.write("We are using the **Diabetes dataset** from sklearn. "
         "We will use only one feature: **BMI** to predict disease progression.")

# --------------------------
# Exploratory Data Analysis
# --------------------------
st.header("1️⃣ Exploratory Data Analysis (EDA)")

# Scatter plot
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(X_bmi, y, alpha=0.6, color="blue")
ax.set_xlabel("BMI")
ax.set_ylabel("Disease Progression")
ax.set_title("Scatter Plot: BMI vs Disease Progression")
st.pyplot(fig)

# Correlation
corr = np.corrcoef(X_bmi["bmi"], y)[0, 1]
st.write(f"🔹 Correlation between BMI and Disease Progression: **{corr:.3f}**")

# --------------------------
# Linear Regression Model
# --------------------------
st.header("2️⃣ Simple Linear Regression")

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_bmi, y, test_size=0.2, random_state=42)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Plot regression line on scatter
fig2, ax2 = plt.subplots(figsize=(6, 4))
ax2.scatter(X_test, y_test, color="blue", alpha=0.6, label="Actual Data")

# Sort for smooth line
X_test_sorted = np.array(sorted(X_test["bmi"])).reshape(-1, 1)
y_pred_line = model.predict(X_test_sorted)

ax2.plot(X_test_sorted, y_pred_line, color="red", linewidth=2, label="Regression Line")
ax2.set_xlabel("BMI")
ax2.set_ylabel("Disease Progression")
ax2.set_title("Regression Line on Test Data")
ax2.legend()
st.pyplot(fig2)

# --------------------------
# Model Evaluation
# --------------------------
st.header("3️⃣ Model Evaluation")

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

st.write(f"🔹 Mean Squared Error (MSE): **{mse:.2f}**")
st.write(f"🔹 R² Score: **{r2:.3f}**")

# Explanation in simple words
if r2 < 0.5:
    st.warning("BMI alone does not explain much of the variation in disease progression. "
               "Other factors (like age, glucose, blood pressure, etc.) are also important.")
else:
    st.success("BMI explains a good portion of disease progression, "
               "but other features may still be helpful.")

st.info("ℹ️ R² score ranges from 0 to 1. For example, R²=0.41 means "
        "BMI explains about 41% of the variation in disease progression.")
```
